As a data scientist our goal is to help users find products that they will love.

Using the data above that contains, products, users and reviews, we will create a recommender system that will predict products based on user past purchases and interests.

This means that when a user logs in to the platform they will receive suggestions based on past purchases and interests.

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['user_id_transform'] = labelencoder.fit_transform(data['user_id'])
```

```python
df = pd.DataFrame({'User ID': data['user_id_transform'].value_counts().index, 'Frequency': data['user_id_transform'].value_counts().values})
```

```python
data['about_product'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#lets create an instance for tfidfvectorizer
tfidf = TfidfVectorizer(stop_words = 'english')

def recommend_products(data, user_id_transform):

  #lets convert the product description into numeric variables using tfidfvectorizer
  data['about_product'] = data['about_product'].fillna('')

  tfidf_matrix = tfidf.fit_transform(data['about_product'])

  #lets get the purchase history of the clients
  user_history = data[data['user_id_transform'] == user_id_transform]

  #lets use cosine to check the similarity between each pair of product description only for the products the user has purchased before
  indices = user_history.index.tolist()

  if indices:
    cosine_sim_user = cosine_similarity(tfidf_matrix[indices], tfidf_matrix)

    #lets create a pandas series with product indices as the index and product names as the values
    products = data.iloc[indices]['product_name']
    indices = pd.Series(products.index, index= products)

    #get the indices and similarity scores of products similar to the ones the user has already purchased
    similarity_scores = list(enumerate(cosine_sim_user[-1]))
    similarity_scores = [(i, score) for (i,score) in similarity_scores if i not in indices]

    #sort the similarity scores in descending order
    similarity_scores = sorted(similarity_scores, key = lambda x:x[1], reverse = True)

    #get the indices of the top 5 most similar products
    top_products = [i[0] for i in similarity_scores[1:6]]
    #get the names of the top 5 similar products
    recommended_products = data.iloc[top_products]['product_name'].tolist()

    #lets get reason for recommendation
    score = [similarity_scores[i][1] for i in range(5) ]

    #lets create a dataframe with the results
    results = pd.DataFrame({'Id Encoded': user_id_transform,
                            'Recommended Product': recommended_products,
                            'Score Recommendation': score })
    return results
  else:
      print('No history purchase found')
      return None
```

```python
recommend_products(data, 1)
```

By analysing the products that each user has purchased, which can identify which categories and features are more attractive to them. This information can be used to optimize product design and marketing  strategies to better target specific groups of customers
